```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
def step_descent(train_x,train_y,m,c,lr):
    m_descent = np.zeros(len(train_x[0]))
    c_descent = 0 
    for i in range(len(train_x)):
        mult_term = 0
        for k in range(len(train_x[0])): 
            mult_term += train_x[i][k] * m[k]
        
        for j in range(len(train_x[0])):
            m_descent[j] += (-2 / len(train_x)) * (train_y[i] - mult_term - c) * (train_x[i][j])
            c_descent += (-2 / len(train_x)) * (train_y[i] - mult_term - c)
        
    new_m = m - lr*m_descent
    new_c = c - lr*c_descent 
    
    return new_m,new_c        
```

```python
def cost(train_x,train_y,m,c):
    
    mult = 0
    cost = 0
    for i in range(len(train_x)):
        mult = 0
        for j in range(len(train_x[0])):
            mult += m[j]*train_x[i][j]
        cost += (train_y[i] - mult - c)**2
    return (cost)
```

```python
def gradient(train_x,train_y,lr = 0.01,num_iter = 10):
    m = np.zeros(len(train_x[0]))
    print("weights shape : ",m.shape)
    c = 0
    for i in range(num_iter):
        m,c = step_descent(train_x,train_y,m,c,lr)
        
        loss = cost(train_x,train_y,m,c)
        print("Interation Number : ",i," : ",loss/len(train_x))
    return m,c
    
```

```python
def predict(test,m,c):
    y = []
    for i in range(len(test)):
        ansy = 0
        for j in range(len(test[0])):
            ansy += m[j]*test[i][j]
        y.append(ansy + c)
    return y
```

```python
train = pd.read_csv("ccpp_x_y_train.csv")
test = np.loadtxt("ccpp_x_test.csv",delimiter = ",")

train_y = pd.DataFrame(train["EP"],columns = [train.columns[-1]])
train_y = np.array(train_y.values)

train_x = train.drop("EP",axis = 1)
train_x = np.array(train_x.values)

scaler = StandardScaler()
scaler.fit(train_x)
train_x = scaler.transform(train_x)
test = scaler.transform(test)
train_x.shape,train_y.shape,test.shape
```

```python
lr = 0.1
num_iter = 500
m,c = gradient(train_x,train_y,lr,num_iter)
```

```python
ypred = predict(test,m,c)
y_pred = np.array(ypred,dtype = float)
np.savetxt("ccpp_self_prac3.csv",y_pred,delimiter = ",",fmt = '%.3f')
```
